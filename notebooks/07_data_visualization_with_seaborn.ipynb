# ðŸŽ¨ 07 â€“ Data Visualization with Seaborn

**Seaborn** is a powerful Python data visualization library built on top of Matplotlib.  
It provides a high-level interface for creating attractive and informative **statistical graphics** with just a few lines of code.

In this notebook, youâ€™ll learn how to:
- Create various chart types such as histograms, boxplots, and scatter plots  
- Visualize relationships and distributions in datasets  
- Customize color palettes and themes  
- Combine Seaborn with Matplotlib for advanced styling  
- Export high-quality visualizations for reports

```python
# Import libraries
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Create images folder if not exists
os.makedirs("images", exist_ok=True)

# Set Seaborn theme
sns.set_theme(style="whitegrid")
```

## ðŸ“Š 1. Load Example Dataset

Seaborn includes several built-in datasets for practice and demonstration.  
We'll use the famous **"tips"** dataset, which contains information about restaurant bills and tips.

```python
# Load Seaborn built-in dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

## ðŸ“ˆ 2. Basic Statistical Plots

Weâ€™ll start by exploring simple charts like histograms, scatter plots, and boxplots.

```python
# Histogram of total bills
plt.figure(figsize=(8,5))
sns.histplot(tips['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bills", fontsize=14)
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.savefig("images/seaborn_hist_totalbill.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Scatter plot of Total Bill vs Tip
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", palette="viridis", s=80)
plt.title("Total Bill vs Tip by Gender and Time", fontsize=14)
plt.savefig("images/seaborn_scatter_tips.png", bbox_inches='tight', dpi=300)
plt.show()
```

## ðŸ§© 3. Categorical Data Visualization

Seaborn makes it easy to visualize categorical variables using bar plots, boxplots, and violin plots.

```python
# Boxplot of tips by day
plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="tip", data=tips, palette="pastel")
plt.title("Boxplot of Tips by Day", fontsize=14)
plt.savefig("images/seaborn_boxplot_tips.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Violin plot to show distribution
plt.figure(figsize=(8,5))
sns.violinplot(x="day", y="tip", hue="sex", data=tips, split=True, palette="muted")
plt.title("Violin Plot of Tips by Day and Gender", fontsize=14)
plt.savefig("images/seaborn_violin_tips.png", bbox_inches='tight', dpi=300)
plt.show()
```

## ðŸŽ¨ 4. Customizing Themes and Color Palettes

Seaborn offers a variety of themes and color palettes to improve the aesthetics of your charts.

```python
# Example of multiple color palettes
palettes = ["deep", "muted", "pastel", "dark", "colorblind"]
fig, axes = plt.subplots(1, len(palettes), figsize=(15,4))

for i, palette in enumerate(palettes):
    sns.histplot(tips["total_bill"], kde=True, color=sns.color_palette(palette)[0], ax=axes[i])
    axes[i].set_title(palette)

plt.tight_layout()
plt.savefig("images/seaborn_color_palettes.png", bbox_inches='tight', dpi=300)
plt.show()
```

## ðŸ§  5. Relationship and Correlation Visualization

Letâ€™s explore how to show **relationships between numerical variables** using pairplots and heatmaps.

```python
# Pairplot to visualize relationships between multiple features
sns.pairplot(tips, hue="sex", palette="coolwarm")
plt.savefig("images/seaborn_pairplot_tips.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
# âœ… Correlation heatmap (fixed version)
plt.figure(figsize=(8,5))

# Select only numeric columns
numeric_tips = tips.select_dtypes(include=["number"])

sns.heatmap(numeric_tips.corr(), annot=True, cmap="YlGnBu", fmt=".2f")

plt.title("Correlation Heatmap (Numerical Columns Only)", fontsize=14)
plt.savefig("images/seaborn_heatmap_corr.png", bbox_inches='tight', dpi=300)
plt.show()
```

## ðŸ’¾ 6. Saving Figures

You can save Seaborn plots as image files (PNG, JPG, SVG, etc.) for reports or documentation.  
Weâ€™ve already demonstrated this with the `plt.savefig()` command in each section.

```python
# Example: Save one more figure
plt.figure(figsize=(8,5))
sns.countplot(x="day", hue="sex", data=tips, palette="coolwarm")
plt.title("Count of Customers by Day and Gender", fontsize=14)
plt.savefig("images/seaborn_countplot.png", bbox_inches='tight', dpi=300)
plt.show()
```
