# üìò 04 ‚Äì Data Visualization with Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib.  
It provides a high-level interface for creating attractive and informative statistical graphics.

In this notebook, we‚Äôll explore:
- Seaborn basics and style settings  
- Scatter, bar, and box plots  
- Pairplots and heatmaps  
- Customizing color palettes and themes  
- Saving visualizations  

```python
# Import necessary libraries
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Ensure the 'images' folder exists
os.makedirs("images", exist_ok=True)

# Set Seaborn style
sns.set(style="whitegrid", palette="pastel")
```

## 1Ô∏è‚É£ Loading a Sample Dataset

Seaborn comes with several built-in datasets.  
Let‚Äôs start by loading the famous **"tips"** dataset.

```python
# Load the 'tips' dataset
tips = sns.load_dataset("tips")
tips.head()
```

## 2Ô∏è‚É£ Scatter Plot

Scatter plots show the relationship between two numeric variables.  
We can also use color (`hue`) and size to represent additional dimensions.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", size="size", palette="viridis", alpha=0.8)
plt.title("Total Bill vs Tip by Day", fontsize=14, fontweight='bold')
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.tight_layout()
plt.show()
```

## 3Ô∏è‚É£ Bar Plot

Bar plots are useful for comparing the average values across categories.

```python
plt.figure(figsize=(7,5))
sns.barplot(data=tips, x="day", y="total_bill", estimator=np.mean, ci=None, palette="coolwarm")
plt.title("Average Total Bill per Day", fontsize=14, fontweight='bold')
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.tight_layout()
plt.show()
```

## 4Ô∏è‚É£ Box Plot

Box plots visualize the distribution of numerical data and highlight outliers.

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="Set2")
plt.title("Box Plot of Total Bill by Day and Gender", fontsize=14, fontweight='bold')
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.tight_layout()
plt.show()
```

## 5Ô∏è‚É£ Violin Plot

Violin plots combine box plots and kernel density estimation to show distribution shape.

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True, palette="muted")
plt.title("Violin Plot ‚Äì Total Bill by Day and Gender", fontsize=14, fontweight='bold')
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.tight_layout()
plt.show()
```

## 6Ô∏è‚É£ Pair Plot

Pair plots visualize pairwise relationships between numeric variables.

```python
sns.pairplot(tips, hue="sex", diag_kind="kde", palette="husl")
plt.suptitle("Pairwise Relationships in Tips Dataset", y=1.02, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 7Ô∏è‚É£ Heatmap

Heatmaps display correlations between numeric variables in a color-coded matrix.

```python
corr = tips.corr(numeric_only=True)

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 8Ô∏è‚É£ Customizing Seaborn Styles

Seaborn allows you to customize themes, palettes, and grid styles easily.

```python
sns.set_style("darkgrid")
sns.set_palette("mako")

plt.figure(figsize=(7,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", s=70)
plt.title("Customized Style Example", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 9Ô∏è‚É£ Saving the Figure

Let‚Äôs save one of our Seaborn figures for future use.

```python
plt.figure(figsize=(7,5))
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Regression Plot ‚Äì Total Bill vs Tip", fontsize=14, fontweight='bold')
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Ensure directory exists and save
os.makedirs("images", exist_ok=True)
plt.savefig("images/seaborn_regression_plot.png", dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Figure saved successfully to 'images/seaborn_regression_plot.png'")
```
