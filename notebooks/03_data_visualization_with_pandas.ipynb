# üìó 03 ‚Äì Data Visualization with Pandas

Pandas has powerful built-in plotting capabilities that make it easy to visualize data directly from DataFrames or Series.  
In this notebook, we‚Äôll learn how to create plots using pandas‚Äô `.plot()` function, and explore different chart types.

We‚Äôll cover:
- Line, bar, and histogram plots  
- Pie and box plots  
- Grouped and stacked plots  
- Correlation heatmaps  
- Customizing and saving plots  

```python
# Import required libraries
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Ensure images directory exists
os.makedirs("images", exist_ok=True)
```

## 1Ô∏è‚É£ Creating a Sample Dataset

Let‚Äôs create a simple dataset to demonstrate pandas plotting capabilities.

```python
# Create sample data
np.random.seed(42)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': np.random.randint(200, 500, 6),
    'Expenses': np.random.randint(150, 400, 6),
    'Profit': np.random.randint(50, 200, 6)
}

df = pd.DataFrame(data)
df
```

## 2Ô∏è‚É£ Line Plot

Pandas integrates seamlessly with Matplotlib. You can quickly visualize trends over time with `.plot(kind='line')`.

```python
df.plot(x='Month', y=['Sales', 'Expenses', 'Profit'], kind='line', figsize=(8,5), marker='o')
plt.title("Company Performance Over Time", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3Ô∏è‚É£ Bar Chart

Bar charts are great for comparing categorical data side by side.

```python
df.plot(x='Month', y=['Sales', 'Expenses'], kind='bar', figsize=(8,5), color=['skyblue', 'orange'])
plt.title("Monthly Sales vs Expenses", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 4Ô∏è‚É£ Histogram

Histograms are useful for understanding the distribution of numeric variables.

```python
df[['Sales', 'Expenses', 'Profit']].plot(kind='hist', bins=10, alpha=0.6, figsize=(8,5))
plt.title("Distribution of Sales, Expenses, and Profit", fontsize=14, fontweight='bold')
plt.xlabel("Value Range")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5Ô∏è‚É£ Pie Chart

Pie charts are suitable for showing proportions of a whole.

```python
df.set_index('Month')['Profit'].plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), startangle=90, colormap='viridis')
plt.title("Profit Distribution per Month", fontsize=14, fontweight='bold')
plt.ylabel("")  # Hide default y-axis label
plt.tight_layout()
plt.show()
```

## 6Ô∏è‚É£ Box Plot

Box plots show the distribution of data and identify potential outliers.

```python
df[['Sales', 'Expenses', 'Profit']].plot(kind='box', figsize=(7,5), grid=True, patch_artist=True)
plt.title("Box Plot of Financial Metrics", fontsize=14, fontweight='bold')
plt.ylabel("Amount ($)")
plt.tight_layout()
plt.show()
```

## 7Ô∏è‚É£ Grouped and Stacked Bar Charts

Pandas allows you to create grouped or stacked bar charts easily.

```python
df.plot(x='Month', y=['Sales', 'Expenses'], kind='bar', stacked=True, figsize=(8,5), color=['#3498db','#e67e22'])
plt.title("Stacked Bar Chart ‚Äì Sales vs Expenses", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 8Ô∏è‚É£ Correlation Heatmap

We can visualize correlations between numerical columns using pandas‚Äô `.corr()` method and Matplotlib‚Äôs `imshow()`.

```python
corr = df[['Sales', 'Expenses', 'Profit']].corr()

plt.figure(figsize=(5,4))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar(label="Correlation")
plt.tight_layout()
plt.show()
```

## 9Ô∏è‚É£ Saving the Figure

You can save pandas-generated figures using Matplotlib‚Äôs `savefig()` function as before.

```python
# Create and save figure
ax = df.plot(x='Month', y='Profit', kind='line', marker='o', figsize=(8,5), color='green')
plt.title("Monthly Profit Trend", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.grid(alpha=0.3)

os.makedirs("images", exist_ok=True)
plt.savefig("images/monthly_profit_trend.png", dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Figure saved successfully to 'images/monthly_profit_trend.png'")
```
