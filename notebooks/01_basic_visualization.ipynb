# üìä 01 ‚Äì Basic Data Visualization with Matplotlib

Data visualization is one of the most important steps in any data science workflow.  
In this notebook, we‚Äôll learn how to use **Matplotlib**, one of the most popular Python libraries for static data visualization.

We'll cover:
- Line plots  
- Bar charts  
- Scatter plots  
- Styling and customization  
- Exporting plots for reports  

```python
# Suppress unnecessary warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Import main visualization and data libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1Ô∏è‚É£ Creating a Simple Line Plot

We'll start by plotting a simple sine wave using NumPy and Matplotlib.

```python
# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='steelblue', linewidth=2, label='Sine Wave')

# Add labels and title
plt.title("Basic Line Plot", fontsize=14, fontweight='bold')
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

## 2Ô∏è‚É£ Plotting Multiple Lines

Sometimes, we need to visualize several variables on the same chart.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Sine', color='royalblue', linewidth=2)
plt.plot(x, y2, label='Cosine', color='orange', linestyle='--', linewidth=2)

plt.title("Sine and Cosine Waves", fontsize=14, fontweight='bold')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## 3Ô∏è‚É£ Creating a Bar Chart

Bar charts are great for comparing categorical data.

```python
# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [12, 19, 7, 15, 9]

plt.figure(figsize=(7, 5))
plt.bar(categories, values, color='teal', alpha=0.8)
plt.title("Bar Chart Example", fontsize=14, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Value")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 4Ô∏è‚É£ Scatter Plot Example

Scatter plots help visualize the relationship between two numerical variables.

```python
# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

plt.figure(figsize=(7, 5))
plt.scatter(x, y, color='darkcyan', edgecolor='black', alpha=0.7)
plt.title("Scatter Plot Example", fontsize=14, fontweight='bold')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(alpha=0.3)
plt.show()
```

## 5Ô∏è‚É£ Customizing Plot Style

Matplotlib provides many built-in styles to make your plots more visually appealing.

```python
# List available styles
plt.style.available[:10]  # Show first 10 styles
```

```python
# Apply a style
plt.style.use('seaborn-v0_8-poster')

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Styled Sine Wave')
plt.title("Styled Matplotlib Plot", fontsize=14)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

## 6Ô∏è‚É£ Saving Figures

You can easily save figures in different formats for reports or presentations.

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Ensure the 'images' folder exists before saving
os.makedirs("images", exist_ok=True)

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create and style the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='darkorange', linewidth=2)
plt.title("Save Plot Example", fontsize=14, fontweight='bold')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(alpha=0.3)

# Save figure to the 'images' directory
plt.savefig("images/sine_wave_plot.png", dpi=300, bbox_inches='tight')

# Display plot
plt.show()

print("‚úÖ Figure saved successfully to 'images/sine_wave_plot.png'")
```
