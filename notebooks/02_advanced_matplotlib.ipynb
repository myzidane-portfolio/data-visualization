# üìò 02 ‚Äì Advanced Visualization Techniques with Matplotlib

In this notebook, we‚Äôll explore **advanced features of Matplotlib** that allow us to create complex and professional plots.

We‚Äôll cover:
- Creating multiple subplots  
- Using twin axes  
- Adding text and annotations  
- Applying custom color maps  
- Exporting publication-quality figures

```python
# Import libraries
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

# Create 'images' folder for saving results
os.makedirs("images", exist_ok=True)
```

## 1Ô∏è‚É£ Creating Multiple Subplots

Matplotlib allows you to create multiple plots in one figure using **subplots()**.  
This is useful when you want to compare several datasets or visualizations side by side.

```python
# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create a figure with 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

axs[0].plot(x, y1, color='royalblue')
axs[0].set_title("Sine Function")

axs[1].plot(x, y2, color='darkorange')
axs[1].set_title("Cosine Function")

axs[2].plot(x, y3, color='seagreen')
axs[2].set_title("Sine √ó Cosine")

for ax in axs:
    ax.set_xlabel("X Values")
    ax.set_ylabel("Y Values")
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 2Ô∏è‚É£ Using Twin Axes

Sometimes, you want to plot two variables that share the same x-axis but have **different scales**.  
Matplotlib allows this using `twinx()` or `twiny()`.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x / 3)

fig, ax1 = plt.subplots(figsize=(8, 5))

# First axis
ax1.plot(x, y1, color='teal', label='Sine Wave')
ax1.set_xlabel("X Axis")
ax1.set_ylabel("Sine", color='teal')
ax1.tick_params(axis='y', labelcolor='teal')

# Second axis (twin)
ax2 = ax1.twinx()
ax2.plot(x, y2, color='darkred', label='Exponential', linestyle='--')
ax2.set_ylabel("Exponential", color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

plt.title("Twin Axes Example", fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.show()
```

## 3Ô∏è‚É£ Adding Annotations and Text

Annotations are essential for highlighting important parts of a plot or providing context.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='steelblue')

# Add annotation
plt.annotate(
    "Peak",
    xy=(np.pi/2, 1),
    xytext=(2, 1.3),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=12,
    fontweight='bold'
)

plt.title("Annotated Sine Wave", fontsize=14, fontweight='bold')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(alpha=0.3)
plt.show()
```

## 4Ô∏è‚É£ Applying Custom Color Maps

Colormaps are useful for continuous or heatmap-style data visualization.

```python
# Generate random data for demonstration
np.random.seed(0)
data = np.random.rand(10, 10)

plt.figure(figsize=(6, 5))
plt.imshow(data, cmap='plasma', interpolation='nearest')
plt.title("Custom Colormap Example (Plasma)", fontsize=14)
plt.colorbar(label='Value')
plt.show()
```

## 5Ô∏è‚É£ Creating a Figure Grid (Advanced Subplots)

Matplotlib allows flexible layout control using `gridspec` or `subplots()` with multiple columns and rows.

```python
# Example of 2x2 grid layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left
axs[0, 0].plot(np.sin(x), color='royalblue')
axs[0, 0].set_title("Sine")

# Top-right
axs[0, 1].plot(np.cos(x), color='darkorange')
axs[0, 1].set_title("Cosine")

# Bottom-left
axs[1, 0].plot(np.tan(x), color='seagreen')
axs[1, 0].set_ylim(-10, 10)
axs[1, 0].set_title("Tangent (Clipped)")

# Bottom-right
axs[1, 1].imshow(data, cmap='viridis')
axs[1, 1].set_title("Heatmap")

plt.tight_layout()
plt.show()
```

## 6Ô∏è‚É£ Exporting Publication-Quality Figures

For academic papers or reports, you can export figures in **high resolution (300+ dpi)** and different formats.

```python
# Create and save a high-quality figure
plt.figure(figsize=(8, 5))
plt.plot(x, np.sin(x) * np.exp(-x/5), color='purple', linewidth=2)
plt.title("Damped Sine Wave", fontsize=14, fontweight='bold')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(alpha=0.3)

# Ensure directory exists
os.makedirs("images", exist_ok=True)

# Save figure in multiple formats
plt.savefig("images/damped_sine_wave.png", dpi=300, bbox_inches='tight')
plt.savefig("images/damped_sine_wave.pdf", dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Figures saved successfully to 'images/' in PNG and PDF formats.")
```
